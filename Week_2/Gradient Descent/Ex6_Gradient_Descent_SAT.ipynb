1. Đọc dữ liệu trên vào dataframe
2. Trực quan hóa dữ liệu theo High School GPA, University GPA
4. X = High School GPA đã chuyển theo định dạng chuẩn, y = University GPA 
5. Với y = mx + b (University GPA = m * High School GPA + b), gọi hàm tính m, b: theta = gradient_descent_2(alpha, X, y, 1000)
6. Từ m, b (m = theta[1], b = theta[0]) => dự đoán University GPA theo m, b
7. Trực quan hóa dữ liệu
9. Với High School GPA là 2.3, 2.8,3.3, 3.8 thì University GPA lần lượt là bao nhiêu?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from sklearn.datasets.samples_generator import make_regression 
from scipy import stats
```

```python
data = pd.read_excel("sat.xls")
data.info()
```

```python
plt.figure(figsize=(12,12))
sns.jointplot(x='high_GPA', y = 'univ_GPA', data=data)
plt.show()
```

```python
from chapter4_lib import *
```

```python
# y = mx + b
m = high_GPA.size
X = np.c_[ np.ones(m), high_GPA] # insert column
y = univ_GPA
alpha = 0.01 # learning rate
theta = gradient_descent_2(alpha, X, y, 1000)
```

```python
X[0:5]
```

```python
print("m = ", theta[1], "b = ", theta[0])
```

```python
for i in range(X.shape[1]):
    univ_GPA_predict = theta[1]* X + theta[0]
```

```python
plt.figure(figsize=(12,8))
plt.xlim(2.0,4.0)
plt.ylim(2.0,4.0)
plt.scatter(X[:,1], univ_GPA, color="green")
plt.plot(X, univ_GPA_predict)
plt.xlabel("high_GPA")
plt.ylabel("univ_GPA")
plt.show()
```

```python
high_GPA_new= np.array([2.3, 2.8,3.3, 3.8])
univ_GPA_new = theta[1]*high_GPA_new + theta[0]
univ_GPA_new
```
